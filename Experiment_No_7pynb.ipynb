```python
# To reverse a number using while loop
n = int(input("Enter the number : "))
rev = 0
while n>0:
  d = n%10                  # To calculate the remainder of the number (e.g when 12/10 , the remainder is 2,which is the last digit of 12)
  rev = (rev*10) + d        #(e.g when num is 12 , 0*10 = 0 and digit is 2  , rev = 2)... i.e now number is 2
  n //= 10                  # To remove the last digit from the number (so 12 becomes 21)

print("The reverse of the number is :" ,rev)
```

```python
# To Check if a number is palindrome or not
n = int(input("Enter the number : "))
temp = n
rev = 0
while n>0:
  d = n%10
  rev = (rev*10) + d
  n //= 10
if(rev==temp):
  print("The number is a palindrome")
else:
  print("The number is not a palindrome")
```

```python
# To Check if a string is palindrome or not

s = "madam"
i,j = 0 , len(s) - 1
is_palindrome = True
while i < j:
    if s[i] != s[j]:
        is_palindrome = False
        break
    i += 1
    j -= 1

if is_palindrome:
    print("Palindrome")
else:
    print("Not a palindrome")
```
